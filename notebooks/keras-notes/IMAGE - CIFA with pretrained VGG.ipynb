## load pretrained vgg16 and vgg19

```python
def build_vgg16(weights_file = None):
    model = Sequential()
    
    model.add(ZeroPadding2D((1,1),input_shape=(3,224,224)))
    model.add(Convolution2D(64, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(64, 3, 3, activation='relu'))
    model.add(MaxPooling2D((2,2), strides=(2,2)))

    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(128, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(128, 3, 3, activation='relu'))
    model.add(MaxPooling2D((2,2), strides=(2,2)))

    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(256, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(256, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(256, 3, 3, activation='relu'))
    model.add(MaxPooling2D((2,2), strides=(2,2)))

    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(512, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(512, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(512, 3, 3, activation='relu'))
    model.add(MaxPooling2D((2,2), strides=(2,2)))

    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(512, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(512, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(512, 3, 3, activation='relu'))
    model.add(MaxPooling2D((2,2), strides=(2,2)))

    model.add(Flatten())
    model.add(Dense(4096, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(4096, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1000, activation='softmax'))
    
    if weights_file is not None:
        model.load_weights(weights_file)
    
    return model

def build_vgg19(weights_file = None):
    model = Sequential()
    
    model.add(ZeroPadding2D((1,1),input_shape=(3,224,224)))
    model.add(Convolution2D(64, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(64, 3, 3, activation='relu'))
    model.add(MaxPooling2D((2,2), strides=(2,2)))

    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(128, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(128, 3, 3, activation='relu'))
    model.add(MaxPooling2D((2,2), strides=(2,2)))

    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(256, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(256, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(256, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(256, 3, 3, activation='relu'))
    model.add(MaxPooling2D((2,2), strides=(2,2)))

    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(512, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(512, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(512, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(512, 3, 3, activation='relu'))
    model.add(MaxPooling2D((2,2), strides=(2,2)))

    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(512, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(512, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(512, 3, 3, activation='relu'))
    model.add(ZeroPadding2D((1,1)))
    model.add(Convolution2D(512, 3, 3, activation='relu'))
    model.add(MaxPooling2D((2,2), strides=(2,2)))

    model.add(Flatten())
    model.add(Dense(4096, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(4096, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1000, activation='softmax'))
    
    if weights_file is not None:
        model.load_weights(weights_file)
    
    return model

def process_img(img):
    #img = transform.resize(img, output_shape=(224, 224))
    img = imresize(img, (224, 224))
    
    ## make it zero mean
    img = util.img_as_ubyte(img).astype("float32")
    img[:,:,0] -= 103.939
    img[:,:,1] -= 116.779
    img[:,:,2] -= 123.68
    
    img = img.transpose((2, 0, 1))
    return img.astype("float32")

class_labels = open("../../data/caffe_ilsvrc12/synset_words.txt").readlines()
```

## Modify the layers of existing VGG
To reuse the existing features, we need to modify the structure (layers) of existing VGG model by removing the top several layers and adding new layers. ***Current keras version (1.0.5) still doesnt have a unified way of doing so:***
 - fchollet (keras administator) suggested to use `model.layers.pop()` and `model.add(XXX)` because calling `pop` alone won't make the output of the model right, and it didnt work with `utils.visualization_plot.plot`
 - use partial weight loading as described in the [code](https://gist.github.com/fchollet/f35fbc80e066a49d65f1688a7e99f069)
 - use separate implementation as discussed [here](https://github.com/fchollet/keras/pull/2418) 
 - we use the [code](https://github.com/fchollet/keras/issues/2640) from joelthchao for the purpose  

```python

## code to pop last layer from model
## cited from https://github.com/fchollet/keras/issues/2640
def pop_layer(model):
    if not model.outputs:
        raise Exception('Sequential model cannot be popped: model is empty.')

    model.layers.pop()
    if not model.layers:
        model.outputs = []
        model.inbound_nodes = []
        model.outbound_nodes = []
    else:
        model.layers[-1].outbound_nodes = []
        model.outputs = [model.layers[-1].output]
    model.built = False
```

```python
from keras.datasets import cifar10
from keras.utils import np_utils
```

```python
(train_x, train_y), (test_x, test_y) = cifar10.load_data()
train_y, test_y = np_utils.to_categorical(train_y), np_utils.to_categorical(test_y)
print train_x[0].min(),train_x.max()
print train_x.shape, test_x.shape
print train_y.shape, test_y.shape
```

```python
from keras.preprocessing.image import ImageDataGenerator

def process_imageset(images):
    images = images.transpose((0, 2, 3, 1))
    images = np.array([imresize(im, (224, 224)) for im in images])
    #print images.shape
    images = images.astype(np.float32)
    images[:, :,:,0] -= 103.939
    images[:, :,:,1] -= 116.779
    images[:, :,:,2] -= 123.68
    images = images.transpose((0, 3, 1, 2))
    return images

def image_generator(images, targets, batch_size = 64):
    """Return batchs
    """
    i = 0
    while True:
        batch_images = images[i:i+batch_size]
        batch_targets = targets[i:i+batch_size]
        batch_images = process_imageset(batch_images)
        yield (batch_images, batch_targets)
        i += batch_size
        if i + batch_size >= images.shape[0]:
            i = 0
```

```python
train_generator = image_generator(train_x, train_y)
test_generator = image_generator(test_x, test_y)
```

```python
## build the vgg model, replace the last layer, free all other layers
cifar_vgg = build_vgg16("../../data/vgg16_keras/vgg16_weights.h5")

## remove the last two layers
# for _ in xrange(5):
#     cifar_vgg.layers.pop()
for _ in xrange(2):
    pop_layer(cifar_vgg)
## set remaining layers as non-trainable    
for layer in cifar_vgg.layers:
    layer.trainable = False
print len(cifar_vgg.layers)
```

```python
## add new trainable layers
cifar_vgg.add(Dense(128, activation="relu"))
cifar_vgg.add(Dense(10, activation="softmax"))
print len(cifar_vgg.layers)
print cifar_vgg.summary()
## small learning rate for deep network
cifar_vgg.compile(loss = "categorical_crossentropy", 
                  optimizer = RMSprop(lr=1e-6), 
                  metrics = ["accuracy"])
```

```python
cifar_vgg.fit_generator(train_generator, samples_per_epoch=train_x.shape[0], nb_epoch=5, 
                        nb_val_samples=64*100, validation_data = test_generator)
```
