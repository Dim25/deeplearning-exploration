### One of the prerequite of deep learning is a lot of data. Several ways have been proposed to make them less depedent of huge amount of data
- Using pretrained network
- Generating data via generative models, e.g., by Generative Adversarial Networks

### This tutorial follows the [blog on Keras](http://blog.keras.io/building-powerful-image-classification-models-using-very-little-data.html) that aims at building deep networks (e.g., CNN) with relatively little data

### Techniques used:
- Use pretrained neural network in Keras
- Fixed a part of NN, and train the rest with new data
- Augmenting dataset (specially images) and use generator as the flexible interface to dataset

```python
import keras
keras.__version__
```

```python
%matplotlib inline
```

```python
from keras import models
from keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

## Use `keras.preprocessing.image.ImageDataGenerator` to augument data

## Keras Image representations are (nchannel, w, h), instead of (w, h, nchannel)

```python
image_augumenter = image.ImageDataGenerator(
    rotation_range = 40,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    ## rescaling is important for both downstream neural network
    ## and display jpeg with plt.imshow
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    fill_mode = "nearest"
)
single_img = image.load_img("../data/cats_dogs/small1000/train/cats/cat.500.jpg", )
single_img = image.img_to_array(single_img)
print single_img.shape
single_img_data = single_img[np.newaxis, :]
print single_img_data.shape

fig, axes = plt.subplots(5, 4, figsize = (4 * 3, 5 * 3))
axes = axes.ravel()
fig.subplots_adjust(hspace = 0.01, wspace=0.05, )

for i, batch in enumerate(image_augumenter.flow(single_img_data, batch_size=1)):
    if i >= 20: break
    axes[i].imshow(batch[0].transpose((1, 2, 0)))
    print batch.shape, batch.min(), batch.max()
```
